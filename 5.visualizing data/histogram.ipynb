```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
n = 1000

data = np.exp(np.random.randn(n)/2)
plt.plot(data,'s')
```

```python
k = 40

plt.hist(data,bins=k)
plt.show()
```

```python
y,x = np.histogram(data,bins=k)

xx = (x[1:]+x[:-1])/2

plt.plot(xx,y)
plt.show()
```
