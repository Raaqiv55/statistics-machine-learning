```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
```

```python
N = 10000
spikeDur = 10
spikeNumA = .01
spikeNumB = .05

spike_tsA = np.zeros(N)
spike_tsB = np.zeros(N)

spiketimeA = np.random.randint(0,N,int(N*spikeNumA))


for spikei in range(len(spiketimeA)):
    
    bnd_pre = int(max(0,spiketimeA[spikei]-spikeDur/2))
    bnd_pst = int(min(N,spiketimeA[spikei]+spikeDur/2))
    
    spike_tsA[bnd_pre:bnd_pst] = 1

spiketimeB = np.random.randint(0,N,int(N*spikeNumB))
spiketimeB[:len(spiketimeA)] = spiketimeA

for spikei in range(len(spiketimeB)):
    
    bnd_pre = int(max(0,spiketimeB[spikei]-spikeDur/2))
    bnd_pst = int(min(N,spiketimeB[spikei]+spikeDur/2))
    
    spike_tsB[bnd_pre:bnd_pst] = 1
    
```

```python
plt.plot(range(N),spike_tsA,range(N),spike_tsB)
plt.ylim([0,1.2])
plt.xlim([2000,2500])
plt.show()
```

```python
probA = sum(spike_tsA==1) / N
probB = np.mean(spike_tsB)

probAB = np.mean(spike_tsA+spike_tsB==2)

print(probA,probB,probAB)
```

```python
pAgivenB = probAB/probB

pBgivenA = probAB/probA

print('P(A) = %g'%probA)
print('P(A|B) = %g'%pAgivenB)
print('P(B) = %g'%probB)
print('P(B|A) = %g'%pBgivenA)
```
