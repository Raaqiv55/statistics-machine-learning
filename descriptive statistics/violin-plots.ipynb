```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
```

```python
n = 1000
thresh = 5

data = np.exp(np.random.randn(n))
data[data>thresh] = thresh + np.random.randn(sum(data>thresh))*.1

plt.plot(data,'+')
```

```python
plt.hist(data,30)
plt.title('Histogram')
plt.show()

plt.violinplot(data)
plt.title('Violin')
plt.show()
```
